```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

def load_tokenizer_and_model(model_name_or_path):
  return GPT2Tokenizer.from_pretrained(model_name_or_path), GPT2LMHeadModel.from_pretrained(model_name_or_path)


def generate(
    model, tok, text,
    do_sample=True, max_length=50, repetition_penalty=5.0,
    top_k=5, top_p=0.95, temperature=1,
    num_beams=None,
    no_repeat_ngram_size=3
    ):
  input_ids = tok.encode(text, return_tensors="pt")
  out = model.generate(
      input_ids,
      max_length=max_length,
      repetition_penalty=repetition_penalty,
      do_sample=do_sample,
      top_k=top_k, top_p=top_p, temperature=temperature,
      num_beams=num_beams, no_repeat_ngram_size=no_repeat_ngram_size
      )
  return list(map(tok.decode, out))
```

```python
import pymorphy2
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import re
from string import punctuation

morph = pymorphy2.MorphAnalyzer()

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('corpus')
```

```python
with open('/content/new_txt_corp.txt') as f:
    text = f.read()
    text = text.replace('\n', ' ')
    print(text)
```

```python
def preprocessing(text):

  text.lower()
  token = word_tokenize(text)
  russian_stopwords = stopwords.words('russian') + [a for a in punctuation]
  token_2 = [word for word in token if word not in russian_stopwords]
  text=re.sub(r'[^\w\s]+|[\d]+', r'', text).strip()
  for ii in range(len(token_2)):
    token_2[ii] = morph.parse(token_2[ii])[0].normal_form
  text=' '.join(token_2)
  text=text.split()
  return(text)
```

```python
text = preprocessing(text)
text
```

```python
affix = []
root = []
```

```python
for i in text:
  current = i.lower()
  var = morph.parse(current)[0]

  for j in range(len(var.lexeme)):
    a = var.lexeme[j].word
    current_root = ''
    current_affix = ''
    flag_add = True

    try:

      for letter_index in range(len(a)):
        if a[letter_index] == current[letter_index]:
          current_root+=a[letter_index]
        else:
          current_affix+=a[letter_index]
          if flag_add:
            if current_root not in root:
              root.append(current_root)
              current_root = ''
            flag_add = False

    except IndexError:

      if current_affix+a[letter_index+1:] not in affix:
        affix.append(current_affix+a[letter_index+1:])
        current_affix = ''


```

```python
root = sorted(root, key=len)
affix = sorted(affix, key=len)
```

```python
import random
```

```python
first_group={'п', 'г', 'т', 'р'}
second_group={'р', 'в', 'з'}
third_group={'в', 'б', 'г'}
fourth_group={'в', 'ф', 'ю', 'х'}
fifth_group={'н', 'к', 'х'}
sixth_group={'е', 'з', 'д', 'к'}
```

```python
generated_words = []

while len(generated_words)<=100:
  #get_index_root = random.randrange(len(root))
  #get_index_affix = random.randrange(len(affix))

  get_index_root = random.randrange(500,1500)
  get_index_affix = random.randrange(500,1500)

  current = root[get_index_root]+affix[get_index_affix]
  cur_int = set(list(current))
  if len(cur_int.intersection(first_group))>2 or len(cur_int.intersection(second_group))>2 or len(cur_int.intersection(third_group))>2 or len(cur_int.intersection(fourth_group))>2 or len(cur_int.intersection(fifth_group))>2 or len(cur_int.intersection(sixth_group))>2:
    generated_words.append(current)

generated_words
```

```python
generated_words2 = sorted(generated_words)
generated_words2
```

```python
#СГ
sg = []
for i in range(100):
  current = soglas[random.randrange(0,21)] + glas[random.randrange(0,10)]
  sg.append(current)

#ГС
gs = []
for i in range(100):
  gs.append(soglas[random.randrange(0,21)]+glas[random.randrange(0,10)])

#СГС
sgs = []
for i in range(100):
  sgs.append(soglas[random.randrange(0,21)]+glas[random.randrange(0,10)]+soglas[random.randrange(0,21)])

#ССГ
ssg = []
for i in range(100):
  ssg.append(soglas[random.randrange(0,21)]+soglas[random.randrange(0,21)]+glas[random.randrange(0,10)])
```

```python
words = []
for i in range(100):
  words.append(sg[random.randrange(0,100)]+sg[random.randrange(0,100)])
  words.append(sg[random.randrange(0,100)]+sgs[random.randrange(0,100)])
  words.append(sgs[random.randrange(0,100)]+gs[random.randrange(0,100)])
  words.append(ssg[random.randrange(0,100)]+gs[random.randrange(0,100)])
```

```python
gen_words = []

for word in words:
  cur_int = set(list(word))
  if len(cur_int.intersection(first_group))>2 or len(cur_int.intersection(second_group))>2 or len(cur_int.intersection(third_group))>2 or len(cur_int.intersection(fourth_group))>2 or len(cur_int.intersection(fifth_group))>2 or len(cur_int.intersection(sixth_group))>2:
    gen_words.append(word)

gen_words
```
